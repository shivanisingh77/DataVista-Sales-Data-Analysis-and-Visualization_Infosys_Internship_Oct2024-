# Deriving year and month from 'order_date' FOR FUTURE ANALYSIS

```python
sales['month_year'] = sales['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
sales['month_year'].sample(10)
```

```python
sales.head()
```

# Grouping sales by month

```python
sales_by_month = sales.groupby('month_year')['sales'].sum().reset_index()
```

# Visualizing sales trend by months

```python
plt.figure(figsize=(15,6))
plt.plot(sales_by_month['month_year'],sales_by_month['sales'])
plt.xticks(rotation='vertical',size=8)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.bar(sales_by_month['month_year'],sales_by_month['sales'])
plt.xticks(rotation='vertical',size=8)
plt.show()
```

# Displaying most selling products

```python
products_sales = pd.DataFrame(sales.groupby('product_name')['sales'].sum())
products_sales = products_sales.sort_values('sales',ascending=False)
```

# Top 10 most sold products

```python
products_sales[:10]
```

```python
products_by_quantity = pd.DataFrame(sales.groupby('product_name')['quantity'].sum())
products_by_quantity_sorted = products_by_quantity.sort_values('quantity',ascending=False)
```

# Top 10 most quantity selling products item

```python
products_by_quantity_sorted[:10]
```

# Visualizing most used shipping mode

```python
plt.figure(figsize=(10,8.5))
sns.countplot(x='ship_mode', hue='ship_mode', data=sales, palette='viridis')
plt.title('Count of Ship Modes')
plt.show()
```

# Distribution of Shiping Modes

```python
ship_mode = pd.DataFrame(sales['ship_mode'].value_counts()).reset_index()
ship_mode = ship_mode.rename({ 'ship_mode': 'shiping_mode'}, axis=1)
ship_mode.head()
```

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(ship_mode['count'], labels = ship_mode['shiping_mode'])
plt.show()
```
